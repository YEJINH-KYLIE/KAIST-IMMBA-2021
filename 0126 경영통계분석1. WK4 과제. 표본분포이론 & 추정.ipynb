```python
# 표본평균 780시간의 z-score 구하기 
zscore = (780-800)/(40/math.sqrt(16))
zscore
```

```python
# 표준정규분포표 그래프 그리는 사용자 함수 
def draw_z_score(x, cond, mu, sigma, title):
    y = stats.norm.pdf(x, mu, sigma)
    z = x[cond]
    plt.plot(x, y)
    plt.fill_between(z, 0, stats.norm.pdf(z, mu, sigma))
    plt.title(title)
    plt.show()
```

```python
stats.norm.cdf(-2)
```

### 문제 1 의 정답

```python
# 3 시그마까지 그래프표시하고, z-score 이하 영역 표시
# 확률도 반환 
x = np.arange(-3,3,0.001)
z0 = zscore
draw_z_score(x, x<z0, 0, 1, 'z < -2')
print("확률은 : ", round(0.022750131948179195,6))
```
